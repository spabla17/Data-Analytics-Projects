# Import Pandas

```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Make a list of missing value types

```python
missing_values=["na", "n/a", "__", "s"]
```

# Dataset added into Dataframe

```python
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values=missing_values)
```

```python
df
```

# Dropping NAN

```python
df = df.dropna()
```

# Column Names

```python
df.columns
```

# Data Type Info

```python
df.dtypes
```

# Add the column for Borough

```python
df['Borough']=df['School DBN'].str[2]
```

Here the Borough is represented by the various letters (M is Manhattan, K is Brooklyn, Q is Queens, R is Staten Island, and X is Bronx.)

```python
df
```

# Converting Data Types to Numeric

```python
Conversion=['Mean Score','Number Scoring Below 65','Percent Scoring Below 65','Number Scoring 65 or Above','Percent Scoring 65 or Above','Number Scoring 80 or Above','Percent Scoring 80 or Above']
```

```python
df[Conversion]=df[Conversion].apply(pd.to_numeric,errors='coerce')
```

```python
df.dtypes
```

# Grouping the data by 'School level'

```python
mean_scores=df.groupby('School Level')['Mean Score'].mean()
```

```python
mean_scores
```

# Filtering for 2015 as per the results above. 

This provides all the exams taken in 2015. 

```python
df2015=df[df['Year']==2015]
```

```python
df2015
```

# Filtering the 2015 data to pull out all high schools only as a point of comparison. 

```python
dfhs=df2015[df2015['School Level']=='High school']
```

```python
dfhs
```

Filtering by Common Core Eng

```python
CCE=dfhs[dfhs['Regents Exam']=='Common Core English']
```

```python
CCE
```

# Brooklyn High School for Law and Technology

```python
Brooklyn=CCE[CCE['School Name']=='Brooklyn High School for Law and Technology']
```

```python
Brooklyn
```

Mean score of Brooklyn HS in 2015 for Common Core English

```python
Brooklyn1=CCE['Mean Score'].mean()
```

```python
Brooklyn1
```

# Comparing Brooklyn Mean Score to that of all Brooklyn High Schools

```python
AllBKHS=CCE[CCE['Borough']=='K']
```

```python
AllBKHS
```

```python
AllBKHS_mean=AllBKHS.groupby(['Regents Exam'])['Mean Score'].mean()
```

```python
AllBKHS_mean
```

# Visualizing the Analysis

Comparing Brooklyn HS for Law and Technology vs all high schools in the Borough of Brooklyn

Mean score for Brooklyn.

```python
Brooklyn1=68.5
```

```python
AllBKHS_mean=67.3
```

```python
plt.bar('Brooklyn High Schools',AllBKHS_mean, color='pink')
plt.bar('Brooklyn HS for Law & Tech',Brooklyn1,color='purple')
plt.xlabel('High Schools',fontsize=15)
plt.ylabel('Mean Score',fontsize=15)
plt.title('Mean Scores for Common Core English Regents',fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

While overall the mean scores are not very different, Brooklyn HS for Law and Technology had a higher mean score as it relates to the Common Core English Regents Exam.
