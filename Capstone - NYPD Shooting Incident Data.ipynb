```python
df_drop= df[['INCIDENT_KEY', 'OCCUR_DATE','BORO','LOCATION_DESC','PERP_AGE_GROUP', 'PERP_SEX', 'PERP_RACE',
       'VIC_AGE_GROUP', 'VIC_SEX', 'VIC_RACE']]
```

```python
#rename columns for ease
```

```python
df_drop = df_drop.rename(columns={'INCIDENT_KEY': 'Incident Id', 'OCCUR_DATE': 'Date', 'BORO': 'Borough'}) 
```

```python
df_drop
```

```python
#Cleaning the data further by droping rows with '(null)' and 'unknown' in the respective categories.
```

```python
df_drop.replace(['(null)', 'UNKNOWN', 'U'], pd.NA, inplace=True)
df2 = df_drop.dropna()
```

```python
df2
```

```python
#Parsing out the data to analyze the shooting by race and gender of the perpetrator
```

```python
df2.hist(column='PERP_RACE', by='PERP_SEX', bins=60)
plt.show()
```

```python
#Based on the data above, black females and males are the main perpetrators within both genders. Black males far exceed the number of black females compartively. 
```

```python
#Now, we analyze the data by Race and Borough
df2.hist(column='Borough', by='PERP_RACE', bins=80)
plt.show()
```

```python
#Based on the visulization above, those of the Black race were involved in the most shootings in Brooklyn. American Indians/Alaskan Natives were involved in the least number of shootings in the Bronx. 
```

```python
#Parsing out the data to analyze the shooting by race and gender of the victim
```

```python
df2.hist(column='VIC_RACE', by='VIC_SEX', bins=60)
plt.show()
```

```python
#Interestingly enough, the victims of the shootings were also generally Blacks within both female and male groups. 
```

```python
#Now, we analyze the data by age group.
```

```python
df2.hist(column='PERP_AGE_GROUP', by='VIC_AGE_GROUP', bins=60)
plt.show()
```

```python
#looking at the plot above, it appears as if most of the victims of the shootings were within the similar age group as the perpetrator, posisbly due to personal issues. Those victims who were less than the age of 18 and those who 65+ were the two age groups whose perpetrators were not generally within the same age group. 
```
