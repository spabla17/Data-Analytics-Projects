# Plot the Comparisons

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

Plots showing how many males vs. females take Diabetes Meds vs. those who do not

```python
plt.figure(figsize=(8,6))
sns.barplot(x='gender',y='count', hue='diabetesMed', data=Cau_diabetes_meds_y_n,color='purple')
plt.title('Comparison of M/F taking Diabetes Medication')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='DiabetesMed')
plt.show()
```

Above we can see that a comparable number of females and males take the diabetes meds. Similarly, from the population tested, a similar amount of males and females do not take diabetes meds.
