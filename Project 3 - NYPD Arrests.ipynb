```python
#Update the dataset to reflect full form of / description of column name
```

```python
df['ARREST_BORO'].mask(df['ARREST_BORO']=="K", 'Brooklyn', inplace=True)
df['ARREST_BORO'].mask(df['ARREST_BORO']=="B", 'Bronx', inplace=True)
df['ARREST_BORO'].mask(df['ARREST_BORO']=="M", 'Manhattan', inplace=True)
df['ARREST_BORO'].mask(df['ARREST_BORO']=="S", 'Staten Island', inplace=True)
df['ARREST_BORO'].mask(df['ARREST_BORO']=="Q", 'Queens', inplace=True)
df['LAW_CAT_CD'].mask(df['LAW_CAT_CD']=="F", 'Felony', inplace=True)
df['LAW_CAT_CD'].mask(df['LAW_CAT_CD']=="V", 'Violation', inplace=True)
df['LAW_CAT_CD'].mask(df['LAW_CAT_CD']=="M", 'Misdemeanor', inplace=True)
df['PERP_SEX'].mask(df['PERP_SEX']=="F", 'Female', inplace=True)
df['PERP_SEX'].mask(df['PERP_SEX']=="M", 'Male', inplace=True)
```

```python
#ensuring the dataset was updated as per the code above.
```

```python
df.head(6)
```

```python
#obtain list of all columns
```

```python
df.columns
```

```python
# Drop columns that are deemed unnecessary
df1= df[['ARREST_KEY', 'OFNS_DESC','LAW_CAT_CD', 'ARREST_BORO', 'AGE_GROUP', 'PERP_SEX', 'PERP_RACE']]
```

```python
df1.head
```

```python
#rename columns for ease
```

```python
df1 = df1.rename(columns={'OFNS_DESC': 'Offense', 'LAW_CAT_CD': 'Felony/Misdemeanor', 'ARREST_BORO': 'Borough', 
                               'AGE_GROUP': 'Age', 'PERP_SEX': 'Sex', 'PERP_RACE': 'Race'}) 
```

```python
df1
```

```python
#Parsing out the data to analyze the arrest by race and gender
```

```python
df1.hist(column='Race', by='Sex', bins=60)
plt.show()
```

```python
#Based on the above visualization, for both females and males, black and white hispanics have the highest arrests.
```

```python
# Cleaning the data further by droping rows with 9, I, null in level of offense or felony/misdemeanor column and "unknown" in the race category:
```

```python
df1.replace(['9', 'null', 'I', 'UNKNOWN'], pd.NA, inplace=True)
df2 = df1.dropna()
```

```python
#Next, we analyze data by amount of felonies, violations, and misdemeanors committed in specific Boroughs.
```

```python
df2.hist(column='Borough', by='Felony/Misdemeanor', bins=10)
plt.show()
```

```python
#Based on the visualization above, the most felonies and misdeameanors occurred in Brooklyn, while number of violations was the highest in Manhattan. 
```

```python
# Now, we analyze the data by Race and Borough
df2.hist(column='Felony/Misdemeanor', by='Race', bins=60)
plt.show()
```

```python
#Overall, irrespective of gender, each race tends to have a high number of misdemeanors, with White hispanics and Blacks leading.
```

```python
#Conclusion Overall
# Based on the above analysis, it is evidence that Blacks and White hispanics have the highest arrest rates. It is important to note that there may be a potential bias related to the number of blacks in the community in comparison to other races in addition to the fact that police is known to target blacks versus other populations without a valid reason. Sometime just being black is enough cause for police to suspect and assume that the black person was the criminal when in fact it might not be the case. 
```
